```python
# Importing necessary libraries
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
# Loading the dataset
train_df = pd.read_csv("../data/raw/data_set_ALL_AML_train.csv")
test_df = pd.read_csv("../data/raw/data_set_ALL_AML_independent.csv")
labels_df = pd.read_csv("../data/raw/actual.csv")
```

```python
# Inspecting the datasets
train_df.head()
```

```python
test_df.head()
```

```python
# Transposing the data
train_df = train_df.set_index("Gene Description")
train_df = train_df.drop("Gene Accession Number", axis=1)

train_df = train_df.T
train_df.head()
```

```python
# Checking the shape
print("Shape is : " , train_df.shape)
```

```python
# Inspecting the labels
labels_df.head()
```

### Data Cleaning and Label Integration

```python
# Check column names
train_df.index[:5]
```

```python
# Inspect the labels
labels_df.head()
```

```python
# Clean labels names
labels_df.columns = ["patient", "cancer"]
```

```python
# Match index type
train_df.index = train_df.index.astype(str)
labels_df["patient"] = labels_df["patient"].astype(str)
```

```python
# Merge labels with training data
df = train_df.merge(labels_df, left_index=True, right_on="patient")
df.head()
```

```python
# Drop unnecessary columns
df = df.drop("patient", axis=1)
```

```python
# Check final shape
print("Final shape:", df.shape)
```

```python
# Encode labels
df["cancer"] = df["cancer"].map({
    "ALL": 0,
    "AML": 1
})
```

```python
# Check final dataset
df.head()
```

### Feature Selection

```python
# Seperate features and target variable
X = df.drop("cancer", axis=1)
y = df["cancer"]
```

```python
# Selecting top features using statistical tests
selector = SelectKBest(score_func=f_classif, k=50) # Select top 50 features
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
selected_features
```

```python
# Create a new dataframe
X_selected = pd.DataFrame(X_new, columns=selected_features)
X_selected.head()
```

```python
# New shape
print("Shape of selected features:", X_selected.shape)
```

### Feature Selection

Due to the high dimensionality of gene expression data, feature selection was applied to reduce the number of genes.

The top 50 genes were selected using a statistical test based on their relationship with the target variable.

This helps improve model performance and reduces overfitting.
